```python
import json # ambil fungsi untuk simpan & baca data
import time # untuk hitung waktu eksekusi

# m = menu kopi

data_menu = [] # list menu kopi

# simpan data ke file
def save_data():
    with open("menu_kopi.json", "w") as f:
        json.dump(data_menu, f)

# muat data dari file
def load_data():
    try:
        with open("menu_kopi.json", "r") as f:
            return json.load(f)
    except:
        return []

# cari menu berdasarkan kode
def search_menu(kode):
    for m in data_menu:
        if m['kode'] == kode:
            return m
    return None

# """" MENU UTAMA """"

def tambah_menu():
    kode = input("Kode menu: ")
    if search_menu(kode):
        print("Menu dengan kode ini sudah ada.")
        return
    nama = input("Nama kopi: ")
    stok = int(input("Stok cup: "))
    harga = int(input("Harga per cup: "))
    data_menu.append({"kode": kode, "nama": nama, "stok": stok, "harga": harga})
    print("Menu ditambahkan.\n")

def lihat_menu():
    if not data_menu:
        print("Belum ada data menu.")
        return
    start = time.time()
    for m in data_menu:
        print(f"{m['kode']} - {m['nama']} (Stok: {m['stok']} cup, Harga: Rp{m['harga']})")
    end = time.time()
    print(f"Waktu tampilkan menu: {end - start:.6f} detik\n")

def edit_menu():
    kode = input("Masukkan kode menu yang ingin diedit: ")
    m = search_menu(kode)
    if m:
        m['stok'] = int(input("Stok baru: "))
        m['harga'] = int(input("Harga baru: "))
        print("Data menu diperbarui.\n")
    else:
        print("Menu tidak ditemukan.\n")

def hapus_menu():
    kode = input("Masukkan kode menu yang ingin dihapus: ")
    m = search_menu(kode)
    if m:
        data_menu.remove(m)
        print("Menu dihapus.\n")
    else:
        print("Menu tidak ditemukan.\n")

def laporan_stok():
    if not data_menu:
        print("Belum ada data.")
        return
    print("Laporan stok kopi (dari terbanyak):")
    start = time.time()
    urut = sorted(data_menu, key=lambda x: x['stok'], reverse=True)
    for m in urut:
        print(f"{m['nama']} - Stok: {m['stok']} cup, Harga: Rp{m['harga']}")
    end = time.time()
    print(f"Waktu proses laporan: {end - start:.6f} detik\n")

def ringkasan_menu():
    if not data_menu:
        print("Belum ada data.")
        return
    total_menu = len(data_menu)
    total_nilai = sum(m['stok'] * m['harga'] for m in data_menu)
    print(f"Total jenis kopi: {total_menu}")
    print(f"Total nilai persediaan: Rp{total_nilai}\n")

def cari_menu():
    kode = input("Masukkan kode menu: ")
    m = search_menu(kode)
    if m:
        print(f"Ditemukan: {m['kode']} - {m['nama']} (Stok: {m['stok']} cup, Harga: Rp{m['harga']})\n")
    else:
        print("Menu tidak ditemukan.\n")

# """" PROGRAM UTAMA """"

data_menu = load_data()

while True:
    print("=== Menu Inventaris Café ===")
    print("1. Tambah Menu Kopi")
    print("2. Lihat Semua Menu")
    print("3. Edit Stok & Harga")
    print("4. Hapus Menu")
    print("5. Laporan Stok")
    print("6. Ringkasan Menu")
    print("7. Cari Menu Kopi")
    print("8. Simpan & Keluar")

    pilihan = input("Pilih menu (1-8): ")
    print()

    if pilihan == '1': tambah_menu()
    elif pilihan == '2': lihat_menu()
    elif pilihan == '3': edit_menu()
    elif pilihan == '4': hapus_menu()
    elif pilihan == '5': laporan_stok()
    elif pilihan == '6': ringkasan_menu()
    elif pilihan == '7': cari_menu()
    elif pilihan == '8':
        save_data()
        print("Data disimpan. Keluar.")
        break
    else:
        print("Pilihan tidak valid.\n")
```
